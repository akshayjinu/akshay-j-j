```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data set
df=pd.read_excel('/content/iris (2).xls')
```

```python
df
```

```python
#Display the columns
print("Columns in the dataset:")
print(df.columns)
```

```python
#Calculating the mean of each column of the dataset.
mean_values = df.mean(numeric_only=True)
print(mean_values)
```

```python
# Check for null values in the dataset
df.isnull().sum()
```

```python
#Perform meaningful visualizations using the dataset
# Visualization 1: Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(10, 15))
df[['SL', 'SW', 'PL', 'PW']].plot(kind='line')
plt.show()
```

```python
column_names = df.columns[:-1]
```

```python
plt.pie(mean_values, labels=column_names, autopct="%1.1f%%")
plt.show()
```

```python
plt.bar(df["SL"],df["SW"])
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```
